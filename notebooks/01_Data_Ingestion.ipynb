```python
master_codes = set(fund_master["amfi_code"].astype(str))
nav_codes = set(nav_history["amfi_code"].astype(str))

missing_in_nav = master_codes - nav_codes
orphan_nav_records = nav_codes - master_codes

print(f"AMFI codes in fund_master: {len(master_codes)}")
print(f"AMFI codes in nav_history: {len(nav_codes)}")
print(f"Missing in nav_history: {len(missing_in_nav)}")
print(f"Orphan NAV codes: {len(orphan_nav_records)}")
```

### Data Quality Summary

No scheme codes are missing from the NAV history dataset, and there are no orphan NAV records that lack corresponding entries in the fund master dataset.

Conclusion

The datasets are fully consistent with respect to AMFI scheme codes. No data quality issues related to scheme code mapping were identified.

This validation confirms that the datasets are suitable for downstream tasks such as NAV trend analysis, return calculations, fund comparisons, and portfolio analytics without requiring additional code reconciliation.
